Pix2Pix expects merged JPEG training images that consist of the 256x256 output image and the 256x256 input image (you can find an example below). In this notebook we will create these training images that consist of the brain scans and associated tumor segmentations.

![](http://i.imgur.com/afSM6qP.jpg)

```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm

# Ground truth
HGG_GT_TXT = '/mnt/naruto/Seminar-Med-Seg2017/BRATS2017/list/HGG_GT.txt'
LGG_GT_TXT = '/mnt/naruto/Seminar-Med-Seg2017/BRATS2017/list/LGG-GT.txt'

# 
HGG_IMG_TXT = '/mnt/naruto/Seminar-Med-Seg2017/BRATS2017/list/HGG-img.txt'
LGG_IMG_TXT = '/mnt/naruto/Seminar-Med-Seg2017/BRATS2017/list/LGG-img.txt'
```

```python
def get_lines(path):
    csv_file = os.path.abspath(path)
    with open(csv_file) as f:
        data_file = f.read().splitlines()
    return [x for x in data_file if len(x) > 0]
```

```python
lgg_set = list(get_lines(LGG_IMG_TXT))
hgg_set = list(get_lines(HGG_IMG_TXT))
lgg_gt_set = list(get_lines(LGG_GT_TXT))
hgg_gt_set = list(get_lines(HGG_GT_TXT))

print('{} lgg images'.format(len(lgg_set)))
print('{} hgg images'.format(len(hgg_set)))
print('{} lgg gt images'.format(len(lgg_gt_set)))
print('{} hgg gt images'.format(len(hgg_gt_set)))
```

```python
def merge_horizontally(path):
    suffix = path.split('_')[-1].split('.')[0]
    segment_path = path.replace(suffix, 'seg')
    combined_path = path.replace(suffix, 'combined')

    assert os.path.isfile(path), path + " is no file"
    assert os.path.isfile(segment_path), segment_path + " is no file"

    images = [Image.open(segment_path), Image.open(path)]

    for img in images:
        img.thumbnail((256, 256), Image.ANTIALIAS)

    total_width, max_height = (512, 256)

    new_im = Image.new('RGB', (total_width, max_height))

    x_offset = 0
    for im in images:
        plt.imshow(im)
        new_im.paste(im, (x_offset,0))
        x_offset += im.size[0]

    #plt.imshow(new_im)
    #plt.show()

    new_im.save(combined_path)
```

```python
path = '/mnt/naruto/Seminar-Med-Seg2017/BRATS2017/LGG/Png-LGG/Brats17_2013_0_1_flair.nx.92.png'
merge_horizontally(path)
```

```python
from multiprocessing import Pool
import time

for file_set in [lgg_set, hgg_set]:
    p = Pool(8)
    max_ = len(file_set)
    with tqdm(total=max_) as pbar:
        for i, _ in tqdm(enumerate(p.imap_unordered(merge_horizontally, file_set))):
            pbar.update()           
    pbar.close()
    p.close()
    p.join()
```

```python
import glob
import shutil
combined_files_lgg = glob.glob("/mnt/naruto/Seminar-Med-Seg2017/BRATS2017/LGG/Png-LGG/*combined*")
combined_files_hgg = glob.glob("/mnt/naruto/Seminar-Med-Seg2017/BRATS2017/HGG/Png-HGG/*combined*")
DATASET_DIR = '/mnt/naruto/Seminar-Med-Seg2017/SegMed/pix2pix-tensorflow/datasets/'

def copy_lgg(path):
    shutil.copy(path, DATASET_DIR + 'brain-lgg/' + path.split('/')[-1])
    
def copy_hgg(path):
    shutil.copy(path, DATASET_DIR + 'brain-hgg/' + path.split('/')[-1])

for (file_set, copy_func) in [(combined_files_lgg, copy_lgg), (combined_files_hgg, copy_hgg)]:
    p = Pool(8)
    max_ = len(file_set)
    with tqdm(total=max_) as pbar:
        for i, _ in tqdm(enumerate(p.imap_unordered(copy_func, file_set))):
            pbar.update()           
    pbar.close()
    p.close()
    p.join()
```
